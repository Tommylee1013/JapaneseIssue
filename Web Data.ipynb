```python
from bs4 import BeautifulSoup
import requests
from janome.tokenizer import Tokenizer
import re

def yahoo_search(keyword, iterator) :
    verbose = ''

    for i in range(iterator) :
        url = "https://search.yahoo.co.jp/search?p="+search+"&ei=UTF-8&x=nl&b="+str(i)+'1'
        header = {"user-agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}
        response = requests.get(url, headers=header)
        response.encoding = response.apparent_encoding

        bs = BeautifulSoup(response.text, 'html.parser')
        bs_ul = bs.find('div')

        for bs_li in bs_ul.find_all('div'):
            bs_text = bs_li.text
            if bs_text != '' and bs_text != '\n' :
                verbose += bs_text

    string_without_newline = re.sub(r"\n[\s0-9'月火水木金土日']", " ", verbose)
    string_without_special_chars = re.sub('[\\s]+', '', string_without_newline)

    # 形態素解析用オブジェクトの生成
    text = Tokenizer()
    txt = string_without_special_chars

    # テキストを一行ごとに処理 - 5
    word_dic = {}
    lines_1 = txt.split("\r\n")
    #print(lines_1)
    #print("\n")
    for line in lines_1:
        malist = text.tokenize(line)
        for w in malist:
            word = w.surface
            ps = w.part_of_speech # 品詞 - 6
            if ps.find("名詞") < 0: continue # 名詞だけをカウント - 7
            if not word in word_dic:
                word_dic[word] = 0
            word_dic[word] += 1

    # よく使われる単語を表示 - 8
    keys = sorted(word_dic.items(), key = lambda x:x[1] , reverse=True)
    tab = ['検索','件','質問','条件','回答','&','キーワード','結果','数','解決','広告','脱字','誤字','指定','ヒント','カテゴリ','専門','ブラウザ','閲覧','プライバシー','問い合わせ','-','キーワード','Q','©',':','指定']

    print(keys)
    for i in range(int(iterator)) :
        if keys[i][0].isalpha() == False :
            keys.pop(i)
        if keys[i][0] in '！＠＃＄％＾＆＊（）：:「」『』’”/(/) /)':
            keys.pop(i)
        if keys[i][0].isascii() :
            keys.pop(i)
        if keys[i][0] in tab :
            keys.pop(i)
        if keys[i][0].isnumeric() == True :
            keys.pop(i)
    for word, cnt in keys[:100]:
        print("{0}({1}) ".format(word,cnt), end="")
    keys = dict(keys[61:])
    return keys
```
